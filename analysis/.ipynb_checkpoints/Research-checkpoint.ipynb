```python
import torch
import torch.nn as nn

class Discriminator(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(np.product(input_dim), 120)
        self.fc2 = nn.Linear(120, 84)
        self.src_out = nn.Linear(84, 1)
        self.class_out = nn.Linear(84, num_classes)

    def forward(self, x):
        x = x.flatten(-2)
        x = F.leaky_relu(self.fc1(x))
        x = F.leaky_relu(self.fc2(x))
        return self.src_out(x), self.class_out(x)

class Generator(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(Generator, self).__init__()
        self.input_dim = input_dim
        self.fc1 = nn.Linear(np.product(input_dim) + 1, 120)
        self.fc2 = nn.Linear(120, 240)
        self.fc3 = nn.Linear(240, 120)
        self.fc4 = nn.Linear(120, np.product(input_dim))

    def forward(self, input_data, target_class):
        x = torch.cat((input_data.flatten(-2), target_class.view(target_class.shape[0], 1).float()), 1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = torch.tanh(self.fc4(x))
        return x.view(-1, *self.input_dim)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import time
import datetime


def train_gan(x, y, num_classes, config=None):
    """
    Trains a StarGAN to build a discriminator that classifies traffic,
    and a generator that modifies traffic to trick the discriminator
    """
    if config is None:
        config = {}
    # Find the input dimensions
    input_dim = x.shape[-2:]
    # Hyperparameters
    batch_size = config.get("batch_size", 16)
    g_lr = config.get("g_lr", 0.0001)
    d_lr = config.get("d_lr", 0.0001)
    beta1 = config.get("beta1", 0.5)
    beta2 = config.get("beta2", 0.999)
    lambda_gp = config.get("lambda_gp", 10)
    lambda_cls = config.get("lambda_cls", 2)
    lambda_rec = config.get("lambda_rec", 10)
    lambda_del = config.get("lambda_del", 10)
    n_critic = config.get("n_critic", 4)
    log_step = config.get("log_step", 100)
    lr_update_step = config.get("lr_update_step", 1000)
    num_iters_decay = config.get("num_iters_decay", 10000)
    num_iters = config.get("num_iters", 1000)
    # Training device
    device = torch.device('cpu')
    # Models
    discriminator = Discriminator(input_dim, num_classes)
    generator = Generator(input_dim, num_classes)
    # Optimizers
    g_optimizer = torch.optim.Adam(generator.parameters(), g_lr, [beta1, beta2])
    d_optimizer = torch.optim.Adam(discriminator.parameters(), d_lr, [beta1, beta2])
    # Convert the training set to tensors
    x_tensor = torch.from_numpy(x)
    y_tensor = torch.from_numpy(y)
    # Create a dataset from the training tensors
    dataset = TensorDataset(x_tensor, y_tensor)
    # Abstract tensor data loading
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    def classification_loss(logit, target):
        return F.cross_entropy(logit, target)
    def gradient_penalty(y, x):
        """Compute gradient penalty: (L2_norm(dy/dx) - 1)**2."""
        weight = torch.ones(y.size()).to(device)
        dydx = torch.autograd.grad(outputs=y,
                                   inputs=x,
                                   grad_outputs=weight,
                                   retain_graph=True,
                                   create_graph=True,
                                   only_inputs=True)[0]
        dydx = dydx.view(dydx.size(0), -1)
        dydx_l2norm = torch.sqrt(torch.sum(dydx**2, dim=1))
        return torch.mean((dydx_l2norm-1)**2)
    # Resets the gradients on our optimizers
    def reset_grad():
        g_optimizer.zero_grad()
        d_optimizer.zero_grad()
    # Updates the learning rate
    def update_lr(g_lr, d_lr):
        """Decay learning rates of the generator and discriminator."""
        for param_group in g_optimizer.param_groups:
            param_group['lr'] = g_lr
        for param_group in d_optimizer.param_groups:
            param_group['lr'] = d_lr
    # Record the time before training starts
    start_time = time.time()
    # Initialize an iterater over the data loader
    data_iter = iter(data_loader)
    for i in range(num_iters):
        # =================================================================================== #
        #                             1. Preprocess input data                                #
        # =================================================================================== #
        # Take a batch from the dataset iterator
        try:
            x_real, label_org = next(data_iter)
        # Restart iteration over the dataset iterator when it runs out of batches
        except StopIteration:
            data_iter = iter(data_loader)
            x_real, label_org = next(data_iter)
        # Generate target domain labels randomly.
        rand_idx = torch.randperm(label_org.size(0))
        label_trg = label_org[rand_idx]

        c_org = label_org.clone()
        c_trg = label_trg.clone()

        x_real = x_real.to(device)           # Input images.
        c_org = c_org.to(device)             # Original domain labels.
        c_trg = c_trg.to(device)             # Target domain labels.
        label_org = label_org.to(device)     # Labels for computing classification loss.
        label_trg = label_trg.to(device)     # Labels for computing classification loss.

        # =================================================================================== #
        #                             2. Train the discriminator                              #
        # =================================================================================== #

        # Compute loss with real images.
        out_src, out_cls = discriminator(x_real)
        d_loss_real = - torch.mean(out_src)
        d_loss_cls = classification_loss(out_cls, label_org)

        # Compute loss with fake images.
        x_fake = generator(x_real, c_trg)
        out_src, out_cls = discriminator(x_fake.detach())
        d_loss_fake = torch.mean(out_src)

        # Compute loss for gradient penalty.
        alpha = torch.rand(x_real.size(0), 1, 1, 1).to(device)
        x_hat = (alpha * x_real.data + (1 - alpha) * x_fake.data).requires_grad_(True)

        out_src, _ = discriminator(x_hat)
        d_loss_gp = gradient_penalty(out_src, x_hat)

        # Backward and optimize.
        d_loss = d_loss_real + d_loss_fake + lambda_cls * d_loss_cls + lambda_gp * d_loss_gp
        reset_grad()
        d_loss.backward()
        d_optimizer.step()

        # Logging.
        loss = {}
        loss['D/loss_real'] = d_loss_real.item()
        loss['D/loss_fake'] = d_loss_fake.item()
        loss['D/loss_cls'] = d_loss_cls.item()
        loss['D/loss_gp'] = d_loss_gp.item()

        # =================================================================================== #
        #                               3. Train the generator                                #
        # =================================================================================== #

        if (i+1) % n_critic == 0:
            # Original-to-target domain.
            x_fake = generator(x_real, c_trg)
            out_src, out_cls = discriminator(x_fake)
            g_loss_fake = - torch.mean(out_src)
            g_loss_cls = classification_loss(out_cls, label_trg)

            # Target-to-original domain.
            x_reconst = generator(x_fake, c_org)
            g_loss_rec = torch.mean(torch.abs(x_real - x_reconst))

            # Modification distance

            g_loss_del = torch.mean(torch.abs(x_fake - x_real))

            # Backward and optimize.
            g_loss = g_loss_fake + lambda_rec * g_loss_rec + lambda_cls * g_loss_cls + lambda_del * g_loss_del
            reset_grad()
            g_loss.backward()
            g_optimizer.step()

            # Logging.
            loss['G/loss_fake'] = g_loss_fake.item()
            loss['G/loss_rec'] = g_loss_rec.item()
            loss['G/loss_cls'] = g_loss_cls.item()
            loss['G/loss_del'] = g_loss_del.item()

        # =================================================================================== #
        #                                 4. Miscellaneous                                    #
        # =================================================================================== #

        # Print out training information.
        if (i+1) % log_step == 0:
            et = time.time() - start_time
            et = str(datetime.timedelta(seconds=et))[:-7]
            log = "Elapsed [{}], Iteration [{}/{}]".format(et, i+1, num_iters)
            for tag, value in loss.items():
                log += ", {}: {:.4f}".format(tag, value)
            print(log)
        # Decay learning rates.
        if (i+1) % lr_update_step == 0 and (i+1) > (num_iters - num_iters_decay):
            g_lr -= (g_lr / float(num_iters_decay))
            d_lr -= (d_lr / float(num_iters_decay))
            update_lr(g_lr, d_lr)
            print ('Decayed learning rates, g_lr: {}, d_lr: {}.'.format(g_lr, d_lr))
    # Return the trained discriminator and generator
    return (discriminator, generator)
def evaluate_discriminator(discriminator, x_tensor, y_true):
    """
    Evaluates the discriminator
    :param x: Tensor input
    :param y: Numpy output
    """
    # Feed the input through the classifier
    _, y_predicted_tensor = discriminator(x_tensor)
    y_predicted = np.argmax(y_predicted_tensor.detach().numpy(), axis=1)
    # Calculate PR-AUC of the classifier's predictions
    pr_auc = average_precision_score(y_true, y_predicted)
    # Calculate a confusion matrix
    confusion_m = confusion_matrix(y_true, y_predicted)
    # Accuracy of the classifier's predictions
    accuracy = accuracy_score(y_true, y_predicted)
    # Collect all metrics into an object
    metrics = {"pr_auc": pr_auc, "confusion_matrix": confusion_m, "accuracy": accuracy}
    # Return metrics
    return metrics
```

```python
class Classifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(Classifier, self).__init__()
        self.fc1 = nn.Linear(np.product(input_dim), 120)
        self.fc2 = nn.Linear(120, 84)
        self.out = nn.Linear(84, 1)

    def forward(self, x):
        x = x.flatten(-2)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return F.sigmoid(self.out(x))
```

```python
from sklearn.metrics import average_precision_score, confusion_matrix, accuracy_score

def train_classifier(x, y, config=None):
    """
    Trains a basic traffic classifier
    """
    if config is None:
        config = {}
    # Find the input dimensions
    input_dim = x.shape[-2:]
    # Hyperparameters
    batch_size = config.get("batch_size", 16)
    lr = config.get("lr", 0.0001)
    beta1 = config.get("beta1", 0.5)
    beta2 = config.get("beta2", 0.999)
    log_step = config.get("log_step", 10)
    num_epochs = config.get("num_epochs", 100)
    # Training device
    device = torch.device('cpu')
    # Models
    classifier = Classifier(input_dim, num_classes)
    # Optimizers
    optimizer = torch.optim.Adam(classifier.parameters(), lr, [beta1, beta2])
    # Loss function
    criterion = nn.BCELoss()
    # Create a dataset from the test training set
    dataset = TensorDataset(x, y)
    # Abstract tensor data loading
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    # Iterate for some number of epochs
    for epoch in range(num_epochs):
        running_loss = 0.0
        for i, data in enumerate(data_loader):
            # get the inputs
            inputs, labels = data
            # zero the parameter gradients
            optimizer.zero_grad()
            # forward + backward + optimize
            outputs = classifier(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            # print statistics
            running_loss += loss.item()
        if epoch % log_step == 0:
            print('[%d] loss: %.3f' %
                  (epoch + 1, running_loss / 2000))
        running_loss = 0.0
    return classifier
def evaluate_classifier(classifier, x_tensor, y_true):
    """
    Evaluates a classifier on a dataset
    :param x: Tensor input
    :param y: Numpy output
    """
    # Feed the input through the classifier
    y_predicted_tensor = classifier(x_tensor)
    y_predicted = y_predicted_tensor.detach().numpy()
    # Calculate PR-AUC of the classifier's predictions
    pr_auc = average_precision_score(y_true, y_predicted)
    # Create a version of the predictions that is absolute
    y_predicted_abs = np.rint(y_predicted)
    # Calculate a confusion matrix
    confusion_m = confusion_matrix(y_true, y_predicted_abs)
    # Accuracy of the classifier's predictions
    accuracy = accuracy_score(y_true, y_predicted_abs)
    # Collect all metrics into an object
    metrics = {"pr_auc": pr_auc, "confusion_matrix": confusion_m, "accuracy": accuracy}
    # Return metrics
    return metrics
```

```python
def print_confusion_matrix(cm, labels):
    for truth, row in enumerate(cm):
        for pred, freq in enumerate(row):
            print("Predicted {} but got {}: {}".format(labels[truth],labels[pred],freq))
```

```python
num_classes = 2
# Create labels for the normal and tor datasets
y_normal = np.repeat(0, normal_data.shape[0])
y_tor= np.repeat(1, tor_data.shape[0])
# Concatenate the normal and tor datasets into inputs and labels
x = np.concatenate((normal_data, tor_data))
y = np.concatenate((y_normal, y_tor))
# Print some info about the dataset
print("Input shape", x.shape)
print("Output shape", y.shape)
```

```python
from sklearn.model_selection import KFold


# Seed the RNGs to ensure consistency
torch.manual_seed(100)
np.random.seed(100)

# Fold configurations
gan_splits = 5
classifier_splits = 3

# Config options for training the GAN
gan_config = {
    'num_iters': 1000,
    'log_step': 100
}
# Config options for training the classifier
classifier_config = {
    'num_epochs': 20,
    'log_step': 5
}


# Store metrics for cross-validated GAN
gan_metrics = []
# Perform k-fold cross validation
for train_idx, test_idx in KFold(n_splits=gan_splits, shuffle=True).split(x):
    
    # Extract the training set from the data
    x_train = x[train_idx]
    y_train = y[train_idx]
    
    # Train the model
    discriminator, generator = train_gan(x_train, y_train, num_classes, gan_config)
    
    # Extract the test set from the data
    x_test = x[test_idx]
    y_test = y[test_idx]
    
    # Store metrics for the cross-validated classifier
    classifier_metrics = []
    # Train and test a classifier
    for cls_train_idx, cls_test_idx in KFold(n_splits=classifier_splits, shuffle=True).split(x_test):
        
        # Extract training data from the test set
        # convert from numpy to torch
        x_cls_train = torch.from_numpy(x_test[cls_train_idx])
        y_cls_train = torch.from_numpy(y_test[cls_train_idx]).float()
        
        # Train a classifier
        classifier = train_classifier(x_cls_train, y_cls_train, classifier_config)
        
        # Extract test data from the test set
        # Convert to tensors
        x_cls_test = torch.from_numpy(x_test[cls_test_idx])
        # Store a copy of the labels for later
        y_cls_test_np = y_test[cls_test_idx]
        y_cls_test = torch.from_numpy(y_cls_test_np)
        # Create a set of labels where the class is flipped for each entry in the test set
        y_cls_test_flipped = 1 - y_cls_test
        # Transform the test set using the generator
        x_cls_test_transformed = generator(x_cls_test, y_cls_test_flipped)
        # Evaluate the classifier's efficacy in determining the original class of the modified traffic
        post_transform_metrics = evaluate_classifier(classifier, x_cls_test_transformed, y_cls_test)
        post_transform_alt_metrics = evaluate_classifier(classifier, x_cls_test_transformed, y_cls_test_flipped.detach().numpy())
        # Accumulate metrics
        cls_fold_metrics = {}
        for key, value in pre_transform_metrics.items():
            cls_fold_metrics["classifier/pre_transform/{}".format(key)] = value
        for key, value in post_transform_metrics.items():
            cls_fold_metrics["classifier/post_transform/{}".format(key)] = value
        for key, value in post_transform_alt_metrics.items():
            cls_fold_metrics["classifier/post_transform/alt/{}".format(key)] = value
        # print(cls_fold_metrics)
        classifier_metrics.append(cls_fold_metrics)
    # Aggregate metrics for this GAN fold
    gan_fold_metrics = {}
    for key in classifier_metrics[0].keys():
        gan_fold_metrics[key] = np.mean([cls_fold[key] for cls_fold in classifier_metrics], axis=0)
    gan_metrics.append(gan_fold_metrics)
metrics = {}
for key in gan_metrics[0].keys():
        metrics[key] = np.mean([gan_fold[key] for gan_fold in gan_metrics], axis=0)
for key, value in metrics.items():
    print("{}:   {}".format(key, value))
```
